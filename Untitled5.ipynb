```python
import os
import sys
import pathlib
import json
import itertools
import codecs
import requests
import copy
import urllib.parse
```

```python
def reigai(name,linelist):
    newlist=[]
    if name == "印旛日本医大":
        for line in linelist:
            if line["code"]!="3010":
                newlist.append(line)
    else:
        for line in linelist:
            if line["code"]!="3010" and line["code"]!="3011":
                newlist.append(line)
    return newlist
        
```

```python
def newhour(hour):
    h=int(hour)
    if h>26:
        h-=24
    return h

def json_make():
    
    #パラメータ
    
    #prefectureCode=都道府県コード
    #corporationName=会社名
    #railName=路線名
    
    date=20201117
    datatype=""
    
    #出力フォルダ名
    expfolder="allstations"
    
    urltype=urllib.parse.quote(datatype)
    
    # 出力フォルダがなければ作成
    exp = pathlib.Path('./'+str(expfolder))
    if exp.exists() is False:
        exp.mkdir()
    
    offset=1
    maxdata=50
    staresult={}
    staresult["Stationlist"]=[]
    
    reigailist=["京成高砂","東松戸","新鎌ケ谷","千葉ニュータウン中央","印旛日本医大"]
    
    while offset < maxdata:
        url="http://api.ekispert.jp/v1/json/station?key=LE_FDhA58aPXdppe&type=train&gcs=wgs84&"+str(datatype)+"&offset="+str(offset)
        r = requests.get(url)
        stadata = json.loads(r.text)
        print(url)
        expfile = codecs.open("./"+str(expfolder)+"/stalist"+str(offset)+".json", 'w', 'utf-8')
        json.dump(stadata, expfile, indent=4, ensure_ascii=False)
        
        if "max" in stadata['ResultSet']:
            maxdata=int(stadata['ResultSet']['max'])
        slist= stadata['ResultSet']['Point']
        

        for station in slist:
            jisho={}
            jisho["Station"]=station["Station"]
            jisho["Prefecture"]=station["Prefecture"]
            jisho["GeoPoint"]=station["GeoPoint"]
            JR="False"
            shitetsu="False"
            stationcode=station["Station"]["code"]
            stationname=station["Station"]["Name"]
            stasum=0
            hourcount=[0]*27

            url="http://api.ekispert.jp/v1/json/operationLine/timetable?key=test_BebQdWgbBbP&date="+str(date)+"&stationCode="+str(stationcode)
            r = requests.get(url)
            data = json.loads(r.text)
            print(url)
            
            if "TimeTable" not in data['ResultSet']:
                jisho["JR"]="False"
                jisho["shitetsu"]="False"
                jisho["Sum"]="0"
                jisho["HourTable"]=[]

                for i in range(3,27):
                    jisho["HourTable"].append({"Hour":str(i),"number":"0"})
                staresult["Stationlist"].append(copy.deepcopy(jisho))
                continue

            resultset={}

            if "code" in data['ResultSet']['TimeTable']:
                linelist= [data['ResultSet']['TimeTable']]
            else:
                linelist = data['ResultSet']['TimeTable']

            resultset["Station"]=stationname
            resultset["Sum"]=[]
            resultset["Linelist"]=[]
            
            if stationname in reigailist:
                linelist = reigai(stationname,linelist)

            result={}
            for line in linelist:
                gokei=int(line['trainCount'])
                stasum+=gokei
                if "ＪＲ" in line['Line']['Name']:JR="True"
                else:shitetsu="True"
                if gokei!=0:
                    Linename=line['Line']['Name']+line['Line']['Direction']+"方面"
                    result["Line"]=Linename
                    result["Linecode"]=line['code']
                    
                    url="http://api.ekispert.jp/v1/json/operationLine/timetable?key=test_BebQdWgbBbP&stationCode="+str(stationcode)+"&date="+str(date)+"&code="+str(line['code'])
                    r = requests.get(url)
                    linedata = json.loads(r.text)
                    result["HourTable"]=[]
                    linehourcount=[0]*27
                    hour = linedata['ResultSet']['TimeTable']['HourTable']
                    
                    if "text" in linedata['ResultSet']['TimeTable']['LineKind']:
                        kindtype=1
                        kindname=[linedata['ResultSet']['TimeTable']['LineKind']['text']]
                    else:
                        kindtype=len(linedata['ResultSet']['TimeTable']['LineKind'])
                        kindname=[""]*kindtype
                        for kind in linedata['ResultSet']['TimeTable']['LineKind']:
                            kindname[int(kind["code"])-1]=kind['text']
                    kindsum=[0]*kindtype
                    hourkind=[[0]*27 for i in range(kindtype)]
                        
                
                    if "MinuteTable" in hour:
                        if "Minute" in hour['MinuteTable']:
                            linehourcount[newhour(hour['Hour'])]=1
                            hourcount[newhour(hour['Hour'])]+=1
                            hourkind[int(hour["MinuteTable"]["Stop"]["kindCode"])-1][newhour(hour['Hour'])]=1
                        else:
                            number=len(hour['MinuteTable'])
                            for minute in hour['MinuteTable']:
                                hourkind[int(minute["Stop"]["kindCode"])-1][newhour(hour['Hour'])]+=1
                            linehourcount[newhour(hour['Hour'])]=number
                            hourcount[newhour(hour['Hour'])]+=number
                    else:
                        for h in hour:
                            if "MinuteTable" in h:
                                number=len(h['MinuteTable'])
                                if "Minute" in h['MinuteTable']:
                                    number-=1
                                    hourkind[int(h["MinuteTable"]["Stop"]["kindCode"])-1][newhour(h['Hour'])]=1
                                else:
                                    for minute in h['MinuteTable']:
                                        hourkind[int(minute["Stop"]["kindCode"])-1][newhour(h['Hour'])]+=1
                                linehourcount[newhour(h['Hour'])]=number
                                hourcount[newhour(h['Hour'])]+=number
                else:
                    Linename=line['Line']['Name']
                    result["Line"]=Linename
                    result["Linecode"]=line['code']
                    result["HourTable"]=[]
                    linehourcount=[0]*27
                    kindtype=1
                    kindname=["普通"]
                    kindsum=[0]*kindtype
                    hourkind=[[0]*27 for i in range(kindtype)]
                for i in range(3,27):
                    kindresult=[]
                    for j in range(kindtype):
                        kindresult.append({"Kindname":kindname[j],"Kindnum":str(hourkind[j][i])})
                    result["HourTable"].append({"Hour":str(i),"number":str(linehourcount[i]),"Kindlist":kindresult})
                resultset["Linelist"].append(copy.deepcopy(result))
                LineColor=line['Line']['Color']
                LineColor=LineColor[0:3]+","+LineColor[3:6]+","+LineColor[6:9]
                kindsum=[]
                for i in range(kindtype):
                    kindsum.append({"Kindname":kindname[i],"Kindnum":str(sum(hourkind[i]))})
                resultset["Sum"].append({"Line":Linename,"number":str(gokei),"LineColor":LineColor,"Kindlist":kindsum})

            jisho["JR"]=JR
            jisho["shitetsu"]=shitetsu
            jisho["Sum"]=str(stasum)
            jisho["HourTable"]=[]

            for i in range(3,27):
                jisho["HourTable"].append({"Hour":str(i),"number":str(hourcount[i])})

            staresult["Stationlist"].append(copy.deepcopy(jisho))

            expfile = codecs.open("./"+str(expfolder)+"/s"+str(stationcode)+".json", 'w', 'utf-8')
            json.dump(resultset, expfile, indent=4, ensure_ascii=False)
        offset+=100
    expfile = codecs.open("./"+str(expfolder)+"/date"+str(date)+".json", 'w', 'utf-8')
    json.dump(staresult, expfile, indent=4, ensure_ascii=False)

json_make()
```
